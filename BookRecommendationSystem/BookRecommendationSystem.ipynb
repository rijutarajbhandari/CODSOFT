```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'rating'
}, inplace=True)
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
rating_with_books=ratings.merge(books, on="ISBN")
```

```python
rating_with_books.head()
```

```python
rating_with_books.shape
```

```python
num_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={
    'rating':'num_of_rating'
}, inplace=True)
```

```python
num_rating.head()
```

```python
final_rating=rating_with_books.merge(num_rating, on ='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
book_pivot=final_rating.pivot_table(columns='user_id', index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
import pickle

pickle.dump(model, open("/content/model.pkl", 'wb'))
pickle.dump(book_names, open("/content/book_names.pkl", 'wb'))
pickle.dump(final_rating, open("/content/final_rating.pkl", 'wb'))
pickle.dump(book_pivot, open("/content/book_pivot.pkl", 'wb'))
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

  for i in range(len(suggestion)):
    books=book_pivot.index[suggestion[i]]
    for j in books:
      print(j)
```

```python
recommend_book('A Bend in the Road')
```
